```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"./2023-Python-Advanced/Week 4/H1.xls")
```

```python
df
```

```python
# 定义 extract_link_after_http 函数
def extract_link_after_http(text):
    index_http = text.find("http")
    if index_http != -1:
        link = text[index_http:]
        return link
    return None

# 假设 df 是你的 DataFrame

# 使用 apply 方法将函数应用到指定列
df['Online Document Address （必填）'] = df['Online Document Address （必填）'].apply(extract_link_after_http)
```
